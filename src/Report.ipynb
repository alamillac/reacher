## DDPG

### One agent

#### Start the environment

We will start solving the environment with only one agent

````python
from ddpg import DDPGAgent
from trainer import Trainer
from environment import Env

env_filename = "../Reacher_One_Linux/Reacher.x86_64"
env = Env(env_filename, train_mode=True)
````

#### Examine the State and Action Spaces

````python
print("Number of agents:", env.num_agents)
print("Number of actions:", env.action_size)
print("States have length:", env.state_size)
````

#### Init Trainer and train the agent

````python
trainer = Trainer(
    save_model_path="../reacher.pth",
    save_checkpoint_path="../reacher_one_checkpoint.pth",
    override_checkpoint=True,
    disable_bar_progress=True
)
agent = DDPGAgent(state_size=env.state_size, action_size=env.action_size, batch_size=64) # batch_size=64 works better for this env
scores = trainer.train_until(env, agent, desired_score=30, consecutive_episodes=100)
````

The values on the score shows the last score, the average of the last five scores and the average of the last 100 scores. [Last, Avg5, Avg100]

#### Plot the scores

````python
trainer.plot_scores(scores)
````

## TD3

---

### One agent

````python
from td3 import TD3Agent
from trainer import Trainer
from environment import Env
import numpy as np

env_filename = "../Reacher_One_Linux/Reacher.x86_64"
env = Env(env_filename, train_mode=True)

low = np.array([-1] * env.action_size)
high = np.array([1] * env.action_size)
action_bounds = (low, high)

print("Number of agents:", env.num_agents)
print("Number of actions:", env.action_size)
print("States have length:", env.state_size)
````

````python
trainer = Trainer(
    save_model_path="../reacher_td3.pth",
    save_checkpoint_path="../reacher_td3_one_checkpoint.pth",
    override_checkpoint=True,
    disable_bar_progress=True
)

agent = TD3Agent(state_size=env.state_size, action_bounds=action_bounds, batch_size=256, n_envs=env.num_agents)
scores = trainer.train_until(env, agent, desired_score=30, consecutive_episodes=100)
````

#### Plot the scores

````python
trainer.plot_scores(scores)
````

### (Optional) Crawler

````python
from td3 import TD3Agent
from trainer import Trainer
from environment import Env
import numpy as np

env_filename = "../Crawler_Linux/Crawler.x86_64"
env = Env(env_filename, train_mode=True)

low = np.array([-1] * env.action_size)
high = np.array([1] * env.action_size)
action_bounds = (low, high)

print("Number of agents:", env.num_agents)
print("Number of actions:", env.action_size)
print("States have length:", env.state_size)

trainer = Trainer(max_t=300, disable_bar_progress=True)
agent = TD3Agent(
    state_size=env.state_size, action_bounds=action_bounds, n_envs=env.num_agents
)

weights_path = "../crawler.pth"
agent.load(weights_path)
scores = trainer.test(env, agent, num_episodes=20)
trainer.plot_scores(scores)
````

## Hyperparameters

---

The hyperparameters used can be found in the agent.py file for ddpg and td3 folder.

### DDPG

```python
BUFFER_SIZE = int(1e5)  # replay buffer size
BATCH_SIZE = 128  # minibatch size
GAMMA = 0.99  # discount factor
TAU = 1e-3  # for soft update of target parameters
LR_ACTOR = 5e-4  # learning rate of the actor
LR_CRITIC = 5e-4  # learning rate of the critic
WEIGHT_DECAY = 0  # L2 weight decay
UPDATE_EVERY = 4  # how often to update the network
MIN_BUFFER_SIZE = 1e4  # minimum buffer size before learning

ACTOR_HIDDEN_LAYER_1 = 256
ACTOR_HIDDEN_LAYER_2 = 128

CRITIC_HIDDEN_LAYER_1 = 256
CRITIC_HIDDEN_LAYER_2 = 128

# Gradient clipping
VALUE_MAX_GRAD_NORM = 1.0

# Prioritized Experience Replay (PER)
PER_ALPHA = 0.6
PER_BETA_START = 0.4
PER_BETA_INCREMENT = 0.001
PER_EPSILON = 1e-5
```
