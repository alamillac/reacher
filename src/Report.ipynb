### One agent

#### Start the environment

We will start solving the environment with only one agent

```python
from ddpg import Agent
from trainer import Trainer
from environment import Env

env_filename = "../Reacher_One_Linux/Reacher.x86_64"
env = Env(env_filename, train_mode=True)
```

#### Examine the State and Action Spaces

```python
print("Number of agents:", env.num_agents)
print("Number of actions:", env.action_size)
print("States have length:", env.state_size)
```

#### Init Trainer and train the agent

```python
trainer = Trainer(
    save_model_path="reacher.pth",
    save_checkpoint_path="reacher_one_checkpoint.pth",
    override_checkpoint=True
)
agent = Agent(state_size=env.state_size, action_size=env.action_size, batch_size=64) # batch_size=64 works better for this env
scores = trainer.train_until(env, agent, desired_score=30, consecutive_episodes=100)
```

#### Plot the scores

```python
trainer.plot_scores(scores)
```

### Many agents

```python
from ddpg import Agent
from trainer import Trainer
from environment import Env

env_filename = "../Reacher_Linux/Reacher.x86_64"
env = Env(env_filename, train_mode=True)

print("Number of agents:", env.num_agents)
print("Number of actions:", env.action_size)
print("States have length:", env.state_size)
```

```python
trainer = Trainer(
    save_model_path="reacher.pth",
    save_checkpoint_path="reacher_many_checkpoint.pth",
    override_checkpoint=True
)
agent = Agent(state_size=env.state_size, action_size=env.action_size, batch_size=128)
scores = trainer.train_until(env, agent, desired_score=30, consecutive_episodes=100)
```

```python
trainer.plot_scores(scores)
```
